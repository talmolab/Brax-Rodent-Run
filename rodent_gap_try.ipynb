```python
from vnl_brax.Rodent_Env_Brax import Rodent
from vnl_brax.base_ant import Walker
```

```python
r = Rodent()
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
import jax
import jax.random as random
import mediapy as media
from tqdm import tqdm

# Generate a random seed
seed = 0

env = r

# Create a random key using the seed
rng_key = jax.random.PRNGKey(seed)
from jax import random

# define the jit reset/step functions
jit_reset = jax.jit(env.reset)
jit_step = jax.jit(env.step)
```

```python
env.action_size
```

```python
# initialize the state
rng = jax.random.PRNGKey(100)
state = jit_reset(rng)
rollout = [state.pipeline_state]
states = [state]

# grab a trajectory
n_steps = 2000
render_every = 1

# might becasue brax does not clip the action to the xml limit in the model
for i in tqdm(range(n_steps)):
  act_rng, rng = jax.random.split(rng)
  ctrl = random.uniform(rng, shape=(30,))
  state = jit_step(state, ctrl)
  rollout.append(state.pipeline_state)
  states.append(state)

media.show_video(env.render(rollout[::render_every], camera='close_profile'), fps=1.0 / env.dt / render_every)
```

```python
media.show_video(env.render(rollout[::render_every], camera=0), fps=1.0 / env.dt / render_every)
```
